```python
from Search_batch import search
def evalulate_sample_generate(LLM,user_input,message_template,reference_list,doc_limit=5,slice_limit=5,threshold=0.4,final_return=10):
    result=search(user_input,doc_limit=doc_limit,slice_limit=slice_limit,threshold=threshold,final_return=final_return)
    evaluate_samples=[]
    for  index,item in enumerate(result):
        query=user_input[index]
        contexts=[i['content'] for i in item]
        response=LLM(
            message_template(query,"\n".join(contexts))
        ).content
        evaluate_samples.append({'user_input':query,'retrieved_contexts':contexts,'response':response,'reference':reference_list[index]})
    return evaluate_samples
        
```

```python
"""
基线RAG配置：

LLM:Doubao-pro-32k

milvus: IV-INF COSINE n_plo 128 文本摘要   n_plo 1024 文本切片

切片原则：512max,按句子切片（\n . \t），保持句子完整性，如果一个句子超过512个字，会将其拆分成256字的循环切片

检索：先摘要搜索限定那些文档里的片段，然后再切片搜索，最后确定返回个数，返回相似度最高的且符合范围的切片 doc_limit:3 splice_limit:10 return:3

提示词：请严格按照以下内容回答用户的问题，如果参考内容中没有相关信息或者无法依据参考信息回答问题，请告知用户。

"""
```

```python
def message_template(query,content):
    return [
        {'role':'system','content':'请严格按照以下内容回答用户的问题，如果参考内容中没有相关信息或者无法依据参考信息回答问题，请告知用户：\n{}'.format(content)},
        {'role':'user','content':query},
    ]
    

```

```python
samples=evalulate_sample_generate(LLM=doubaoAI,message_template=message_template,user_input=user_input,reference_list=reference,doc_limit=3,slice_limit=10,final_return=3)
```

```python
import os
config_path=os.path.join('RAG','config.yaml')
import yaml 
with open(config_path  ,'r') as file:
    config=yaml.safe_load(file)
os.environ['OPENAI_API_KEY']=config['doubaoAI']['api_key']
os.environ['OPENAI_API_BASE']=config['doubaoAI']['base_url']
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
evaluator_llm =LangchainLLMWrapper(ChatOpenAI(model=config['doubaoAI']['ep_point']))

```

```python
embeddings = OpenAIEmbeddings(model=config['doubaoembedding']['ep_point'])
evaluator_embeddings =LangchainEmbeddingsWrapper(embeddings)
```

```python
from ragas import evaluate
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    context_relevancy,
    context_recall,
    context_precision,
    
)
result = evaluate(
    dataset=samples,
    metrics=[
        context_precision,
        faithfulness,
        answer_relevancy,
        context_recall,
    ],
    llm=evaluator_llm,
    embeddings=evaluator_embeddings,
)
result_df=result.to_pandas()
```

```python
from ragas.metrics import LLMContextRecall, Faithfulness, FactualCorrectness,ContextEntityRecall,LLMContextPrecisionWithReference,AnswerCorrectness
from ragas import evaluate
metrics = [
    LLMContextRecall(llm=evaluator_llm), #通过使用带注释的答案和检索到的上下文估计 TP 和 FN 来估计上下文回忆。
    #FactualCorrectness(llm=evaluator_llm), 
    ContextEntityRecall(llm=evaluator_llm), #上下文实体回忆，根据事实和上下文中存在的实体计算召回率。令 CN 为上下文中存在的实体集，GN 为事实中存在的实体集。如果这个数量是 1，我们可以说检索机制已经检索到涵盖基本事实中存在的所有实体的上下文
    LLMContextPrecisionWithReference(llm=evaluator_llm), #查看模型选择上下文的优先级是否与上下文的排序一致。
    #Faithfulness(llm=evaluator_llm), #检索到的上下文是否与事实的真实性一致。
    AnswerCorrectness(llm=evaluator_llm), #检索到的上下文与生成的答案和事实的相关性
    
]
results = evaluate(dataset=dataset, metrics=metrics)
df = results.to_pandas()
df.head()
```
