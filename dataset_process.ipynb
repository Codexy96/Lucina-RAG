```python
#读取CSV文件
import pandas as pd
df=pd.read_csv('dataset.csv')
```

```python
# 转化为json数据
import json
import pandas as pd

json_data = []
max_len=0
for index, row in df.iterrows():
    # 过滤掉为NaN的搜索结果
    search_results = [row[f'搜索结果{i}'] for i in range(1, 16) if pd.notna(row[f'搜索结果{i}'])]
    for i in range(len(search_results)):
        record = {
        'industry': [row['行业']],
        'position': [row['岗位']],
        'keyword': [row['搜索关键词']],
        'source': [row['搜索来源']],
        'text': search_results[i],
        'image': row['图片'].split(',') if pd.notna(row['图片']) else [],
    }
        json_data.append(record)



```

```python
json_data[0]
```

```python
import hashlib
def generate_hash_id(text):
    hasher=hashlib.sha256()
    hasher.update(text.encode('utf-8'))
    return hasher.hexdigest()
```

```python
#去重并合并
#去重并合并
json_dict={}
for item in json_data:
    if generate_hash_id(item['text']) not in json_dict:
        json_dict[generate_hash_id(item['text'])]=item
    else:
        for key in json_dict[generate_hash_id(item['text'])]:
            if key in item and key!='text':
                json_dict[generate_hash_id(item['text'])][key].extend(item[key])


```

```python
json_dict['0001b2b202246d02a04ef9b3bd44b2876dbb62f8585761452d0822cf22270693']
```

```python
for  id,item in json_dict.items():
    for key in item.keys():
            if key!= "text":
                 json_dict[id][key] = " ".join(list(set(json_dict[id][key])))
```

```python
json_dict['0001b2b202246d02a04ef9b3bd44b2876dbb62f8585761452d0822cf22270693']
```

```python
#组合成新的json_data
new_json_data =[]
for key, value in json_dict.items():
    new_json_data.append(value)
```
